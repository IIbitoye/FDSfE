```python
pl.imshow(happiness[['Happiness Score', 'Standard Error',
       'Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual', 'year' ]].corr(),)
pl.xticks([0,2,4,6,8], ['Happiness Score', 'Standard Error',
       'Economy(GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual', 'year'],rotation= 70)
pl.yticks([0,2,4,6,8], ['Happiness Score', 'Standard Error',
       'Economy(GDP per Capita)', 'Family', 'Health (Life Expectancy)',
       'Freedom', 'Trust (Government Corruption)', 'Generosity',
       'Dystopia Residual', 'year'],rotation= 45)
pl.colorbar();
```

This diagram is a graph of the correlation between each variable. The yellow spots represent a perfect correlation, a coreelation of 1.
 From the diagram, It is obvious that the variables have a perfect correlation when compared to themselves, as shown with the yellow diagonal.
 We see an anti-corrrelation in few boxes where the correlation is negative and displays a deep purple colour.
