```python
import pandas as pd
```

```python
bank = pd.read_csv("https://github.com/fedhere/FDSfE_FBianco/raw/main/data/bank_term_deposit.csv")
```

```python
bank.info
```

```python
bank
```

```python
bank["y"].unique()
```

```python
bank.hist(figsize=(10,10));
```

```python
cols=["age", "balance", "duration","marital"]
features=bank[cols]
target=bank["y"]
```

```python
features
```

```python
features["marital"].unique()
```

ONE-HOT Ending

```python
pd.get_dummies(features["marital"])
```

```python
features.join(pd.get_dummies(features["marital"])).drop(columns="marital")
```

```python
features=features.join(pd.get_dummies(features["marital"])).drop(columns="marital")
```

```python
features
```

Min-Max Normalization
(array-min(array))/max(array)-min(array)

```python
((features-features.min())/(features.max()-features.min())).hist(figsize=(10,10));
```

```python
features=((features-features.min())/(features.max()-features.min()))
```

```python
features
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg= LogisticRegression()
```

```python
logreg.fit(features, target)
```

```python
logreg.coef_
```

```python
logreg.predict(features)
```

```python
prediction=logreg.predict(features)
```

```python
(prediction==target).sum()/len(target)
```

confusion matrix model evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target,prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(target, prediction)
```

```python
conf=confusion_matrix(target, prediction)
```

```python
import seaborn as sns
import pylab as plt
sns.heatmap(conf, cmap="Blues", annot=True, fmt="d")
plt.xlabel("Preicted")
plt.ylabel("Actual")
```
